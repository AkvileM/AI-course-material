```python
from typing import List

class Pratimas:
    def __init__ (self, name: str, kiekis: int) -> None:
        self._name = name
        self._kiekis = kiekis

    @property
    def name(self) -> str:
        return self._name

    @property
    def kiekis(self) -> int:
        return self._kiekis

    def _rpr__(self) -> str:
        return f"{self._name} {self._kiekis}"


def skaityti_pratimus(file_name: str) -> List[Pratimas]:
    pratimai: List[Pratimas] = []
    with open(file_name, 'r') as f:
        pratimu_kiekis = int(f.readline())
        for n in range(pratimu_kiekis):
            pavadinimas, kiekis = f.readline().split()
            pratimai.append(Pratimas(pavadinimas, int(kiekis)))
    return pratimai

pratimai: List[Pratimas] = skaityti_pratimus("C:/Users/akvil/OneDrive/Desktop/Dirbtinio intelekto studijos/U2.txt")
#print(set([str(pratimas.name) for pratimas in pratimai]))
for pratimas in list(set([pratimas.name for pratimas in pratimai])):
    #print(pratimas)
    galutinis_pratimu_kiekis = 0
    for saraso_pratimas in pratimai:
        if saraso_pratimas.name == pratimas:
            galutinis_pratimu_kiekis += saraso_pratimas.kiekis
    with  open("C:/Users/akvil/OneDrive/Desktop/Dirbtinio intelekto studijos/U2rez.txt", 'w') as f:
        for saraso_pratimas in pratimai:
            f.write(f"{pratimas}, {galutinis_pratimu_kiekis} \n")

```

```python
from typing import Tuple

from git import BlockingLockFile
from traitlets import Bool
class Avis:
    def __init__(self, vardas:str, dnr:str) -> None:
        self._vardas = vardas
        self._dnr = dnr
        self._sutapimo_koef = 0

    def sutapimas(self, avis) -> int:
        sutapimo_koef = 0
        for i, raide in enumerate(self._dnr):
            if raide == avis.dnr[i]:
                sutapimo_koef += 1
        return sutapimo_koef

    
    @property
    def vardas(self) -> str:
        return self._vardas

    @property
    def dnr(self) -> str:
        return self._dnr

    @property
    def sutapimo_koef(self) -> int:
        return self._sutapimo_koef

    @sutapimo_koef.setter
    def sutapimo_koef(self, koeficientas: int) -> None:
        self._sutapimo_koef = koeficientas

    def __eq(self, __o) -> bool:
        return self.sutapimo_koef == __o.sutapimo_koef and self.vardas ==__o.vardas

    def __lt__(self, __o) -> bool:
        if self._sutapimo_koef == __o.sutapimo_koef:
            return self.vardas > __o.vardas
        return self.sutapimo_koef < __o.sutapimo_koef 

def skaityti(file_name: str) -> Tuple[List[Avis], Avis]:
    avys: List[Avis] = []
    
    with open(file_name, "r") as f:
        f.readline()
        tiriamosios_avies_indeksas = int(f.readline())
        for line in f.readline():
            for i, line in enumerate(f.readlines()):
                vardas, dnr = line.split()
                if i+1 == tiriamosios_avies_indeksas:
                    tiriamoji_avis = Avis(vardas, dnr)
                else:
                    avys.append(Avis(vardas, dnr))
    return avys, tiriamoji_avis
avys, tiriamoji_avis = skaityti("U3.txt")

for avis in avys:
    sutapimo_koef = avis.sutapimas(tiriamoji_avis)
    avis.sutapimo_koef = sutapimo_koef

avys.sort(reverse = True)
with open("U3rez.txt", "w") as f:
    for avis in avys:
        f.write(f"{avis.vardas}, {avis.sutapimo_koef} \n")

avys[0] < avys[1]


```

```python
from typing import List
from typing import Dict
class Savivaldybe:
    def __init__(self, pavadinimas: str, mokyklos: int) -> None:
        self._pavadinimas = pavadinimas
        self._mokyklos = mokyklos

 
    @property
    def pavadinimas(self) -> str:
        return self._pavadinimas
    
    @property
    def mokyklos(self) -> int:
        return self._mokyklos

class Apskritis:
    def __init__(self, pavadinimas: str) -> None:
        self._pavadinimas = pavadinimas
        self._savivaldybes: List[Savivaldybe] = []
    
    @property
    def savivaldybiu_kiekis(self) -> int:
        return len(self._savivaldybes)

    @property
    def max_mokyklu_kiekis(self) -> int:
        return max([savivaldybe.mokyklos for savivaldybe in self._savivaldybes])

    def prideti_savivaldybe(self, savivaldybe: Savivaldybe) -> None:
        self._savivaldybes.append(savivaldybe)

    @property
    def pavadinimas(self) -> str:
        return self._pavadinimas

def skaityti_duomenis(failas: str) -> List[Apskritis]:
    apskriciu_sarasas: Dict[str, Apskritis] = {}
    with open(failas, "r") as f:
        savivaldybiu_kiekis = int(f.readline())
        for i in range(savivaldybiu_kiekis):
            savivaldybes_pavadinimas, mokyklu_skaicius = f.readline().split()
            savivaldybe = Savivaldybe(savivaldybes_pavadinimas, int(mokyklu_skaicius))
            apskrities_pavadinimas = f.readline().replace("\n","")
            if apskrities_pavadinimas in apskriciu_sarasas:
                apskriciu_sarasas[apskrities_pavadinimas].prideti_savivaldybe(savivaldybe)
            else:
#                print(f"Apskritis {apskrities_pavadinimas} dar nera {apskriciu_sarasas}")
                apskriciu_sarasas[apskrities_pavadinimas] = Apskritis(apskrities_pavadinimas)
                apskriciu_sarasas[apskrities_pavadinimas].prideti_savivaldybe(savivaldybe)
    return list(apskriciu_sarasas.values())

apskritys = skaityti_duomenis("U4.txt")
for apskritis in apskritys:
    print(apskritis.pavadinimas, apskritis.savivaldybiu_kiekis, apskritis.max_mokyklu_kiekis) 
```

```python
from datetime import datetime
from typing import Tuple, List

from IPython.utils.text import string

class Slidininkas:
    def __init__(self, vardas: str, pavarde: str)->None:
        self._vardas = vardas
        self._pavarde = pavarde
        self._starto_laikas: datetime = None
        self._finisho_laikas: datetime = None

    @property
    def vardas(self) -> str:
        return self._vardas

    @property
    def pavarde(self) -> str:
        return self._pavarde

    @property
    def starto_laikas(self) ->datetime:
        return self._starto_laikas

    @property
    def finisho_laikas(self) ->datetime:
        return self._finisho_laikas

    @property
    def vardas_pavarde(self)-> str:
        return f"{self._vardas} {self._pavarde}"

    @starto_laikas.setter
    def starto_laikas(self, laikas: datetime) -> None:
        self._starto_laikas = laikas

    @finisho_laikas.setter
    def finisho_laikas(self, laikas: datetime) -> None:
        self._finisho_laikas = laikas

    #funkcija esanti klaseje, kuri neiima vidiniu atributu
    @staticmethod
    def konvertuoti_laika(laikas: str) -> datetime:
       # print(f"*{laikas}*")
        return datetime.strptime(laikas.replace(" ",":"), "%H:%M:%S")

    def slidininko_trukme(self) ->int:
        return (self._finisho_laikas - self._starto_laikas).total_seconds()

    def __eq__(self, __o: object) -> bool:
        return self.slidininko_trukme() == __o.slidininko_trukme() and self.vardas_pavarde == __o.vardas_pavarde

    def __lt__(self, __o: object) ->bool:
        #return self.slidininko_trukme() == __o.slidininko_trukme() and self.vardas_pavarde < __o.vardas_pavarde
        if self.slidininko_trukme() == __o.slidininko_trukme():
            return self.vardas_pavarde < __o.vardas_pavarde

        return self.slidininko_trukme() < __o.slidininko_trukme()

    @staticmethod
    def formatuoti_laika(laikas:int) -> str:
        return f"{int(laikas//60)} {int(laikas%60)}"



def istraukti_duomenis(eilute: str) -> Tuple[str, str, str]:
    vardas_pavarde = eilute[:20]
    laikas = eilute[20:]
    pavarde = vardas_pavarde.split()[-1]
    vardas =  " ".join(vardas_pavarde.split()[:-1])
   # print(vardas, pavarde, laikas)
    return vardas, pavarde, laikas

def skaityti_faila(pavadinimas) -> List[Slidininkas]:
    slidininkai: List[Slidininkas] = []
    with open(pavadinimas, "r") as f:
        startavusiu_slidininku_sk = int(f.readline())
        for _ in range(startavusiu_slidininku_sk):
            eilute = f.readline().rstrip()
            vardas, pavarde, laikas = istraukti_duomenis(eilute)
            slidininkas = Slidininkas(vardas, pavarde)
            slidininkas.starto_laikas = Slidininkas.konvertuoti_laika(laikas)
            slidininkai.append(slidininkas)

        finisavusiu_slidininku_sk = int(f.readline())
        for _ in range(finisavusiu_slidininku_sk):
            eilute = f.readline().rstrip()
            vardas, pavarde, laikas = istraukti_duomenis(eilute)
            for f_slidininkas in slidininkai:
                if f_slidininkas.vardas == vardas and f_slidininkas.pavarde == pavarde:
                    f_slidininkas.finisho_laikas = Slidininkas.konvertuoti_laika(laikas)
                    break
        return slidininkai

slidininkai = skaityti_faila("C:/Users/akvil/OneDrive/Desktop/Dirbtinio intelekto studijos/U5.txt")
#for slidininkas in slidininkai:
 #   print(slidininkas.vardas, slidininkas.pavarde, slidininkas.starto_laikas, slidininkas.finisho_laikas)

finisave_slidininkai = [slidininkas for slidininkas in slidininkai if slidininkas.finisho_laikas ]   

for slidininkas in sorted(finisave_slidininkai):
   print(slidininkas.vardas, slidininkas.pavarde, Slidininkas.formatuoti_laika(slidininkas.slidininko_trukme()))

```
